```python
import matplotlib,nltk
from matplotlib import pyplot as plt
from wordcloud import WordCloud
from nltk import word_tokenize
from nltk.corpus import webtext
```

```python
def getText(txtFileName):
    file = open(txtFileName, 'r')
    return file.read()
```

```python
txt1 = getText('simple_corpus.txt')
```

```python
print("Word Tokenizer 출력 :",word_tokenize(txt1))
```

```python
fdist = nltk.FreqDist(word_tokenize(txt1))
```

```python
clean_words = dict([(m,n) for m,n in fdist.items() if len(m) >3])

wordcloud = WordCloud().generate_from_frequencies(clean_words)

plt.figure(figsize=(12,9))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
